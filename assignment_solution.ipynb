# SOLUTION

In the DFS technique in chapter 6.3, the longer a vertex stays in the stack the earlier it appears in the topological sorting. Vertices that finish last, appear first in the sorting.

In Khan, the first vertex output is one that depends on no other vertices (a *source* vertex, by definition). Khan is essentially a greedy peeling of the graph from the front (from a source vertex).

Both algoritms are dual views of the same structure.

DFS finishes a vertex (ie releases it from the stack) only after all its outgoing edges have been processed. In Khan, a vertex is ready when all its prerequisites have been removed. In other words:

- DFS: I won’t finish vertex $u$ until all its descendants are done.
- Khan: I won’t start vertex $u$ until all its ancestors are gone.

These are time-reversed statements of the same dependency logic.
